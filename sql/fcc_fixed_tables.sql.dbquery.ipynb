```sql
CREATE TABLE IF NOT EXISTS fcc_fixed.fact_geo
(
    id BIGINT GENERATED ALWAYS AS IDENTITY (START WITH 1 INCREMENT BY 1),
    geography_id STRING NOT NULL,
    area_data_type STRING,
    geography_type STRING NOT NULL,
    geography_desc STRING NOT NULL,
    PRIMARY KEY(id, geography_id)
);

CREATE TABLE IF NOT EXISTS fcc_fixed.dim_speed
(
    id BIGINT GENERATED ALWAYS AS IDENTITY (START WITH 1 INCREMENT BY 1),
    geography_id STRING NOT NULL,
    fact_geo_id BIGINT NOT NULL,
    speed_02_02 DECIMAL,
    speed_10_1 DECIMAL,
    speed_25_3 DECIMAL,
    speed_100_20 DECIMAL,
    speed_250_25 DECIMAL,
    speed_1000_100 DECIMAL,
    CONSTRAINT fk_fact_geo_dim_speed FOREIGN KEY (fact_geo_id, geography_id) REFERENCES fcc_fixed.fact_geo(id, geography_id),
    PRIMARY KEY(id)
);

CREATE TABLE IF NOT EXISTS fcc_fixed.dim_technology
(
    id BIGINT GENERATED ALWAYS AS IDENTITY (START WITH 1 INCREMENT BY 1),
    geography_id STRING NOT NULL,
    fact_geo_id BIGINT NOT NULL,
    area_data_type STRING,
    biz_res STRING,
    technology STRING,
    total_units INT,
    CONSTRAINT fk_fact_geo_dim_technology FOREIGN KEY (fact_geo_id, geography_id) REFERENCES fcc_fixed.fact_geo(id, geography_id),
    PRIMARY KEY(id)
);
```
