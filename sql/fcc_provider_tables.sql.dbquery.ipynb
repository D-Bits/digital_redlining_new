```sql
CREATE TABLE IF NOT EXISTS fcc_provider.fact_provider_meta
(
    id BIGINT GENERATED ALWAYS AS IDENTITY (START WITH 1 INCREMENT BY 1),
    provider_id NUMERIC NOT NULL,
    frn BIGINT,
    holding_company VARCHAR(100),
    PRIMARY KEY(id, provider_id)
);

CREATE TABLE IF NOT EXISTS fcc_provider.dim_mobile_provider
(
    id BIGINT GENERATED ALWAYS AS IDENTITY (START WITH 1 INCREMENT BY 1),
    provider_id NUMERIC NOT NULL,
    holding_company VARCHAR(100),
    technology_code INT NOT NULL,
    technology_code_desc VARCHAR(250),
    area_stationary INT,
    area_invehicle INT,
    PRIMARY KEY(provider_id),
    FOREIGN KEY (provider_id) REFERENCES fcc_provider.fact_provider_meta(provider_id)
);

CREATE TABLE IF NOT EXISTS fcc_provider.dim_fixed_provider
(
    id BIGINT GENERATED ALWAYS AS IDENTITY (START WITH 1 INCREMENT BY 1),
    provider_id NUMERIC NOT NULL,
    holding_company VARCHAR(100),
    technology_code INT NOT NULL,
    technology_code_desc VARCHAR(250),
    location_count_res INT,
    unit_count_res INT,
    location_count_bus INT,
    unit_count_bus INT,
    PRIMARY KEY(provider_id),
    FOREIGN KEY (provider_id) REFERENCES fcc_provider.fact_provider_meta(provider_id)
);

```
